```python
import pandas as pd
import numpy as np

# Create a copy to work with
df_decoded = df_filtered.copy()

print("Starting data decoding...")

# ============================================================================
# DECODE ALL COLUMNS TO MEANINGFUL TEXT
# ============================================================================

# TARGET VARIABLES - Heart Disease (1=Yes, 2=No, 7=Don't know, 9=Refused)
yes_no_dict = {1.0: 'Yes', 2.0: 'No', 7.0: 'Unknown', 9.0: 'Refused'}

df_decoded['heart_attack'] = df_decoded['heart_attack'].map(yes_no_dict)
df_decoded['coronary_heart_disease'] = df_decoded['coronary_heart_disease'].map(yes_no_dict)
df_decoded['stroke'] = df_decoded['stroke'].map(yes_no_dict)

# Heart Disease Computed (1=No, 2=Yes)
df_decoded['heart_disease_computed'] = df_decoded['heart_disease_computed'].map({
    1.0: 'No', 2.0: 'Yes'
})

# SEX (1=Male, 2=Female)
df_decoded['sex'] = df_decoded['sex'].map({
    1.0: 'Male', 2.0: 'Female'
})

# AGE GROUPS
df_decoded['age_group_5yr'] = df_decoded['age_group_5yr'].map({
    1.0: '18-24', 2.0: '25-29', 3.0: '30-34', 4.0: '35-39',
    5.0: '40-44', 6.0: '45-49', 7.0: '50-54', 8.0: '55-59',
    9.0: '60-64', 10.0: '65-69', 11.0: '70-74', 12.0: '75-79',
    13.0: '80+', 14.0: 'Unknown'
})

df_decoded['age_group'] = df_decoded['age_group'].map({
    1.0: '18-24', 2.0: '25-34', 3.0: '35-44',
    4.0: '45-54', 5.0: '55-64', 6.0: '65+'
})

# RACE/ETHNICITY
df_decoded['race_ethnicity'] = df_decoded['race_ethnicity'].map({
    1.0: 'White', 2.0: 'Black', 3.0: 'Asian',
    4.0: 'Native American/Alaska Native', 5.0: 'Hispanic', 6.0: 'Other'
})

# EDUCATION LEVEL
df_decoded['education_level'] = df_decoded['education_level'].map({
    1.0: 'Did not graduate High School',
    2.0: 'High School Graduate',
    3.0: 'Some College',
    4.0: 'College Graduate',
    9.0: 'Unknown'
})

# INCOME CATEGORY (appears to have more categories than expected)
df_decoded['income_category'] = df_decoded['income_category'].map({
    1.0: 'Less than $15,000',
    2.0: '$15,000 to <$25,000',
    3.0: '$25,000 to <$35,000',
    4.0: '$35,000 to <$50,000',
    5.0: '$50,000 to <$100,000',
    6.0: '$100,000 to <$200,000',
    7.0: '$200,000+',
    9.0: 'Unknown'
})

# MARITAL STATUS
df_decoded['marital_status'] = df_decoded['marital_status'].map({
    1.0: 'Married',
    2.0: 'Divorced',
    3.0: 'Widowed',
    4.0: 'Separated',
    5.0: 'Never Married',
    6.0: 'Unmarried Couple',
    9.0: 'Refused'
})

# VETERAN STATUS
df_decoded['veteran_status'] = df_decoded['veteran_status'].map(yes_no_dict)

# EMPLOYMENT STATUS
df_decoded['employment_status'] = df_decoded['employment_status'].map({
    1.0: 'Employed for wages',
    2.0: 'Self-employed',
    3.0: 'Out of work <1 year',
    4.0: 'Out of work 1+ years',
    5.0: 'Homemaker',
    6.0: 'Student',
    7.0: 'Retired',
    8.0: 'Unable to work',
    9.0: 'Refused'
})

# GENERAL HEALTH
df_decoded['general_health'] = df_decoded['general_health'].map({
    1.0: 'Excellent',
    2.0: 'Very Good',
    3.0: 'Good',
    4.0: 'Fair',
    5.0: 'Poor',
    7.0: 'Unknown',
    9.0: 'Refused'
})

# HEALTH STATUS COMPUTED
df_decoded['health_status_computed'] = df_decoded['health_status_computed'].map({
    1.0: 'Good or Better', 2.0: 'Fair or Poor', 9.0: 'Unknown'
})

# 14+ UNHEALTHY DAYS
unhealthy_14_dict = {1.0: 'No', 2.0: 'Yes', 3.0: 'None', 9.0: 'Unknown'}
df_decoded['physical_unhealthy_14plus_days'] = df_decoded['physical_unhealthy_14plus_days'].map(unhealthy_14_dict)
df_decoded['mental_unhealthy_14plus_days'] = df_decoded['mental_unhealthy_14plus_days'].map(unhealthy_14_dict)

# DIABETES STATUS
df_decoded['diabetes_status'] = df_decoded['diabetes_status'].map({
    1.0: 'Yes',
    2.0: 'Yes, during pregnancy only',
    3.0: 'No',
    4.0: 'Pre-diabetes/Borderline',
    7.0: 'Unknown',
    9.0: 'Refused'
})

# CHRONIC CONDITIONS (Yes/No)
chronic_conditions = ['has_arthritis', 'has_copd', 'has_kidney_disease',
                     'has_skin_cancer', 'has_other_cancer', 'has_depression',
                     'ever_had_asthma']
for col in chronic_conditions:
    df_decoded[col] = df_decoded[col].map(yes_no_dict)

# BMI CATEGORY
df_decoded['bmi_category'] = df_decoded['bmi_category'].map({
    1.0: 'Underweight',
    2.0: 'Normal Weight',
    3.0: 'Overweight',
    4.0: 'Obese'
})

# OVERWEIGHT/OBESE
df_decoded['overweight_obese'] = df_decoded['overweight_obese'].map({
    1.0: 'No', 2.0: 'Yes', 9.0: 'Unknown'
})

# BMI - divide by 100 to get actual BMI
df_decoded['bmi'] = df_decoded['bmi'] / 100

# WEIGHT - divide by 100 to get actual kg
df_decoded['weight_kg'] = df_decoded['weight_kg'] / 100

# LIFESTYLE BEHAVIORS
df_decoded['exercise_past_30days'] = df_decoded['exercise_past_30days'].map(yes_no_dict)
df_decoded['physical_activity'] = df_decoded['physical_activity'].map({
    1.0: 'Yes', 2.0: 'No', 9.0: 'Unknown'
})

df_decoded['smoked_100_cigarettes'] = df_decoded['smoked_100_cigarettes'].map(yes_no_dict)

df_decoded['current_smoking_frequency'] = df_decoded['current_smoking_frequency'].map({
    1.0: 'Every day',
    2.0: 'Some days',
    3.0: 'Not at all',
    7.0: 'Unknown',
    9.0: 'Refused'
})

df_decoded['smoking_status'] = df_decoded['smoking_status'].map({
    1.0: 'Current smoker - every day',
    2.0: 'Current smoker - some days',
    3.0: 'Former smoker',
    4.0: 'Never smoked',
    9.0: 'Unknown'
})

df_decoded['current_smoker'] = df_decoded['current_smoker'].map({
    1.0: 'No', 2.0: 'Yes', 9.0: 'Unknown'
})

# ALCOHOL
df_decoded['binge_drinking'] = df_decoded['binge_drinking'].map({
    1.0: 'No', 2.0: 'Yes', 9.0: 'Unknown'
})

df_decoded['heavy_drinking'] = df_decoded['heavy_drinking'].map({
    1.0: 'No', 2.0: 'Yes', 9.0: 'Unknown'
})

# Drinks per week - divide by 100 to get actual number
df_decoded['drinks_per_week'] = df_decoded['drinks_per_week'] / 100

# HEALTHCARE ACCESS
df_decoded['primary_insurance'] = df_decoded['primary_insurance'].map({
    1.0: 'Employer',
    2.0: 'Purchased directly',
    3.0: 'Medicare',
    4.0: 'Medigap',
    5.0: 'Medicaid',
    6.0: 'CHAMPUS/TRICARE/VA',
    7.0: 'Indian Health Service',
    8.0: 'State program',
    9.0: 'Other',
    10.0: 'No coverage',
    77.0: 'Unknown',
    88.0: 'No coverage',
    99.0: 'Refused'
})

df_decoded['has_healthcare_coverage'] = df_decoded['has_healthcare_coverage'].map({
    1.0: 'Yes', 2.0: 'No', 9.0: 'Unknown'
})

df_decoded['has_personal_doctor'] = df_decoded['has_personal_doctor'].map({
    1.0: 'Yes, only one',
    2.0: 'More than one',
    3.0: 'No',
    7.0: 'Unknown',
    9.0: 'Refused'
})

df_decoded['could_not_afford_care'] = df_decoded['could_not_afford_care'].map(yes_no_dict)

df_decoded['last_checkup'] = df_decoded['last_checkup'].map({
    1.0: 'Within past year',
    2.0: '1-2 years ago',
    3.0: '2-5 years ago',
    4.0: '5+ years ago',
    7.0: 'Unknown',
    8.0: 'Never',
    9.0: 'Refused'
})

# DISABILITY/FUNCTIONING
disability_cols = ['difficulty_walking', 'difficulty_dressing', 'difficulty_errands',
                  'difficulty_concentrating', 'hearing_difficulty', 'vision_difficulty']
for col in disability_cols:
    df_decoded[col] = df_decoded[col].map(yes_no_dict)

# SOCIAL DETERMINANTS
df_decoded['worried_unemployment'] = df_decoded['worried_unemployment'].map(yes_no_dict)

df_decoded['food_insecurity'] = df_decoded['food_insecurity'].map({
    1.0: 'Always',
    2.0: 'Usually',
    3.0: 'Sometimes',
    4.0: 'Rarely',
    5.0: 'Never',
    7.0: 'Unknown',
    9.0: 'Refused'
})

df_decoded['difficulty_paying_bills'] = df_decoded['difficulty_paying_bills'].map(yes_no_dict)
df_decoded['utilities_shut_off'] = df_decoded['utilities_shut_off'].map(yes_no_dict)
df_decoded['lack_transportation'] = df_decoded['lack_transportation'].map(yes_no_dict)

df_decoded['stress_level'] = df_decoded['stress_level'].map({
    1.0: 'Not at all',
    2.0: 'A little bit',
    3.0: 'Moderately',
    4.0: 'Quite a bit',
    5.0: 'A great deal',
    7.0: 'Unknown',
    9.0: 'Refused'
})

df_decoded['life_satisfaction'] = df_decoded['life_satisfaction'].map({
    1.0: 'Very satisfied',
    2.0: 'Satisfied',
    3.0: 'Dissatisfied',
    4.0: 'Very dissatisfied',
    7.0: 'Unknown',
    9.0: 'Refused'
})

df_decoded['emotional_support'] = df_decoded['emotional_support'].map({
    1.0: 'Always',
    2.0: 'Usually',
    3.0: 'Sometimes',
    4.0: 'Rarely',
    5.0: 'Never',
    7.0: 'Unknown',
    9.0: 'Refused'
})

# PREVENTIVE CARE
df_decoded['flu_shot'] = df_decoded['flu_shot'].map(yes_no_dict)
df_decoded['pneumonia_vaccine'] = df_decoded['pneumonia_vaccine'].map(yes_no_dict)

print("✓ All columns decoded successfully!")

# Save the decoded dataset
df_decoded.to_csv('LLCP2022_heart_disease_decoded.csv', index=False)
print(f"✓ Saved as 'LLCP2022_heart_disease_decoded.csv'")
print(f"  Shape: {df_decoded.shape}")

# Display sample
print("\n" + "="*80)
print("SAMPLE OF DECODED DATA (First 10 rows, key columns)")
print("="*80)
sample_cols = ['age_group_5yr', 'sex', 'general_health', 'bmi_category',
               'heart_attack', 'coronary_heart_disease', 'stroke',
               'diabetes_status', 'smoking_status']
print(df_decoded[sample_cols].head(10))

Show target variable distribution
print("\n" + "="*80)
print("TARGET VARIABLE DISTRIBUTION")
print("="*80)

print("\n1. Heart Attack:")
print(df_decoded['heart_attack'].value_counts())
print(f"   Prevalence: {(df_decoded['heart_attack'] == 'Yes').sum() / len(df_decoded) * 100:.2f}%")

print("\n2. Coronary Heart Disease:")
print(df_decoded['coronary_heart# _disease'].value_counts())
print(f"   Prevalence: {(df_decoded['coronary_heart_disease'] == 'Yes').sum() / len(df_decoded) * 100:.2f}%")

print("\n3. Stroke:")
print(df_decoded['stroke'].value_counts())
print(f"   Prevalence: {(df_decoded['stroke'] == 'Yes').sum() / len(df_decoded) * 100:.2f}%")

print("\n" + "="*80)
print("✓ DATA READY FOR ANALYSIS!")
print("="*80)
```
